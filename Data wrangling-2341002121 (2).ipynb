```python
df1=pd.DataFrame({'key':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df2=pd.DataFrame({'key':['a','b','d'],
                  'data2':range(3)})
print(df1)
print(df2)
```

```python
pd.merge(df1,df2)# it will consider the data which have common elements in both dataframe.
```

```python
pd.merge(df1,df2,on='key') # we go for on when the columns name are same
```

```python
pd.merge(df1,df2,how='inner')# it will merge the data which are common elements in both dataframe
```

```python
pd.merge(df1,df2,how='outer')# it will merge every data wwhich are present in both dataframe
```

```python
pd.merge(df1,df2,how='left')# it will merge the data on the basis of first dataframe with second dataframe
```

```python
pd.merge(df1,df2,how='right')# it will merge on the basis of 2nd dataframe.
```

```python
df3=pd.DataFrame({'lkey':['b','b','a','c','a','a','b'],
                   'data1':range(7)})
df4=pd.DataFrame({'rkey':['a','b','d'],
                  'data2':range(3)})
pd.merge(df3,df4,left_on='lkey',right_on='rkey')# we use left on and and right on because of differnent column names
```
